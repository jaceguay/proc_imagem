```python
import geemap
import ee
```

```python
# Initialize the Earth Engine module
ee.Initialize()
# ee.Authenticate()
```

```python
# Define your area of interest
longitude = -48.6290
latitude = -26.9574
point = ee.Geometry.Point(longitude, latitude)

# Set start and end date for image retrieval
start_date = '2020-01-01'
end_date = '2023-12-31'

# Filter the collection for the specified date and region
collection = (ee.ImageCollection('LANDSAT/LC08/C01/T1')
              .filterDate(start_date, end_date)
              .filterBounds(point))
```

```python
collection
```

```python
# Get the first image from the collection
#image = collection.first()

index = 4
image = ee.Image(collection.toList(collection.size()).get(index))
```

```python
# Display the map
Map = geemap.Map()
Map.centerObject(point, 16)
Map.addLayer(image, {'bands': ['B4', 'B3', 'B2'], 'max': 5000}, 'Landsat 8')
Map.addLayerControl()
Map
```
