```python
import cv2
from ultralytics import YOLO
import numpy as np
import pandas as pd
```

```python
# captura a ser analizada e modelo a ser utilizado

cap = cv2.VideoCapture("/Users/jac/Documents/temp/IMG_3339.MOV")

model = YOLO("yolov8m.pt")
```

```python
# quadro a quadro da captura

while True:
    ret, frame = cap.read()
    if not ret:
        break

    quadro = cap.get(cv2.CAP_PROP_POS_FRAMES)

    if quadro % 60 == 0:
        results = model(frame, device="mps")
        result = results[0]
        bboxes = np.array(result.boxes.xyxy.cpu(), dtype="int")
        classes = np.array(result.boxes.cls.cpu(), dtype="int")
        for cls, bbox in zip(classes, bboxes):
            (x, y, x2, y2) = bbox
            cv2.rectangle(frame, (x, y), (x2, y2), (0, 0, 225), 2)
            cv2.putText(frame, str(cls), (x, y - 5),
                        cv2.FONT_HERSHEY_PLAIN, 2, (0, 0, 225), 2)

    # salvar cada quadro na pasta capturas
    #    cv2.imwrite(
    #        "capturas/" + str(cap.get(cv2.CAP_PROP_POS_FRAMES)) + ".jpg", frame)

    # columnas: coordenada x, y do centróide, classe, quadro e segundo
        for cls, bbox in zip(classes, bboxes):
            (x, y, x2, y2) = bbox
            x_centroide = (x + x2) / 2
            y_centroide = (y + y2) / 2
            df = pd.DataFrame(
                {
                    "x_centroide": [str(int(x_centroide))],
                    "y_centroide": [str(int(y_centroide))],
                    "classe": [str(int(cls))],
                    "quadro": [str(int(quadro))],
                    "segundo": [str(int(quadro / 60))],
                }
            )
            df.to_csv(f"centroids/{str(int(quadro))}.csv",
                      header=False, mode="a", index=False)

cap.release()
cv2.destroyAllWindows()
```
